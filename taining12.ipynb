Описание
Дан массив X положительных целых чисел. Его элементы должны быть преобразованы путем выполнения над ними следующей операции столько раз, сколько необходимо:

if X[i] > X[j] then X[i] = X[i] - X[j]

Если больше преобразований невозможно, вернуть его сумму («наименьшую возможную сумму»).

Например, последовательное преобразование элементов входа X = [6, 9, 21] подробно описано ниже:

X_1 = [6, 9, 12] # -> X_1[2] = X[2] - X[1] = 21 - 9
X_2 = [6, 9, 6]  # -> X_2[2] = X_1[2] - X_1[0] = 12 - 6
X_3 = [6, 3, 6]  # -> X_3[1] = X_2[1] - X_2[0] = 9 - 6
X_4 = [6, 3, 3]  # -> X_4[2] = X_3[2] - X_3[1] = 6 - 3
X_5 = [3, 3, 3]  # -> X_5[1] = X_4[0] - X_4[1] = 6 - 3
Возвращаемый результат представляет собой сумму окончательного преобразования (здесь 9).

Пример
solution([6, 9, 21]) #-> 9
Этапы решения:
[6, 9, 12] #-> X[2] = 21 - 9
[6, 9, 6] #-> X[2] = 12 - 6
[6, 3, 6] #-> X[1] = 9 - 6
[6, 3, 3] #-> X[2] = 6 - 3
[3, 3, 3] #-> X[1] = 6 - 3
Дополнительные примечания:
Существуют тесты производительности, состоящие из очень больших чисел и массивов размером не менее 30000. Пожалуйста, напишите эффективный алгоритм для предотвращения тайм-аута.

```python
from math import gcd
from functools import reduce

def solution(X):
    return len(X) * reduce(gcd, X)
```
