Например, давайте рассмотрим следующую задачу. Представьте, что у вас есть пирамида, построенная из чисел, как эта:

   /3/
  \7\ 4 
 2 \4\ 6 
8 5 \9\ 3
Вот и задача...
Допустим, что «скольжение вниз» — это максимальная сумма последовательных чисел от вершины до основания пирамиды. Как вы видите, самое длинное «скольжение вниз» — это3 + 7 + 4 + 9 = 23

Ваша задача — написать функцию, которая принимает представление пирамиды в качестве аргумента и возвращает ее наибольшее 'скольжение вниз' . Например:

* With the input [[3], [7, 4], [2, 4, 6], [8, 5, 9, 3]]
* Your function should return 23.
Кстати...
Мои тесты включают в себя некоторые необычайно высокие пирамиды, так что, как вы можете догадаться, метод грубой силы — плохая идея, если только у вас нет нескольких столетий, которые можно потратить впустую. Вы должны придумать что-то более умное, чем это.

(c) Это задание представляет собой лирическую версию задачи 18 и/или задачи 67 из ProjectEuler .

```python
def longest_slide_down(pyramid):
    # Копируем пирамиду, чтобы не изменять исходные данные
    dp = [row[:] for row in pyramid]

    # Начинаем с предпоследнего ряда
    for row in range(len(dp) - 2, -1, -1):
        for col in range(len(dp[row])):
            # Обновляем текущий элемент как максимальную сумму из нижних элементов
            dp[row][col] += max(dp[row + 1][col], dp[row + 1][col + 1])

    # Вершина пирамиды содержит максимальную сумму пути
    return dp[0][0]

# Тестируем
pyramid = [
    [3],
    [7, 4],
    [2, 4, 6],
    [8, 5, 9, 3]
]

print(longest_slide_down(pyramid))  # 23
```
